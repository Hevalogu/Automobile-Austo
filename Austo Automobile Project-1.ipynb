```python
def outlier(col):
 sorted(col)
 Q1,Q3=col.quantile([0.25,0.75])
 IQR=Q3-Q1
 lr= Q1-(1.5 * IQR)
 ur= Q3+(1.5 * IQR)
 return lr, ur 
```

```python
lr1,ur1=outlier(Dataset['Total_salary'])
Dataset['Total_salary']=np.where(Dataset['Total_salary']>ur1,ur1,Dataset['Total_salary'])
Dataset['Total_salary']=np.where(Dataset['Total_salary']<lr1,lr1,Dataset['Total_salary'])
```

```python
lr2,ur2=outlier(Dataset['No_of_Dependents'])
Dataset['No_of_Dependents']=np.where(Dataset['No_of_Dependents']>ur1,ur1,Dataset['No_of_Dependents'])
Dataset['No_of_Dependents']=np.where(Dataset['No_of_Dependents']<lr1,lr1,Dataset['No_of_Dependents'])
```

```python
plt.figure(figsize= (15,10))
sns.boxplot(data=Dataset)
```
